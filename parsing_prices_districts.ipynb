```python
# Parse the HTML content
soup = BeautifulSoup(html_content, 'html.parser')

# Find all divs with district and price info
district_price_divs = soup.find_all('div', style=True)

# Extract data into lists
districts_prices = {}


for div in district_price_divs:
    # Extract district name (text before <strong>)
    district = div.contents[0].strip()

    # Extract price (inside <strong> tags)
    price = float(div.strong.text.strip())

    # Make a dictionary - districts as keys, prices as values
    districts_prices[district] = price
```
